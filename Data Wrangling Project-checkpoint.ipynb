```python
#Check the number of Tweet ids
tweet_ids = df.tweet_id.values
print("Number of tweet_ids: " + str(len(tweet_ids)) + "\n")
```

```python
#Extract tweets from the twitter API with a loop
tweet_data = []
#Tweets that found are saved in this list:
tweet_id_found = []
#Tweets that are not found are saved in this list:
tweet_id_missing = []
for tweet_id in df['tweet_id']:
    try:
        data = api.get_status(tweet_id, 
                              tweet_mode='extended',
                              wait_on_rate_limit = True,
                              wait_on_rate_limit_notify = True)
        tweet_data.append(data)
        tweet_id_success.append(tweet_id)
    except Exception as e:
        tweet_id_missing.append(tweet_id)
```

```python
#Query Twitter API wiht the tweet_ids from the WeRateDogs twitter archive  for each tweet's. They are saved in jSON format
count = 0
fails_dict = {}
start = timer()
# Each tweet is saved and returned JSON as a new line in a .txt file
with open('tweet_json.txt', 'w') as file:
    for tweet_id in tweet_ids:
        count += 1
        print(str(count) + ": " + str(tweet_id))
        try:
            tweet = api.get_status(tweet_id, tweet_mode='extended')
            print("Success")
            json.dump(tweet._json, file)
            file.write('\n')
        except Exception as e:
            print("Fail")
            fails_dict[tweet_id] = e
            pass
end = timer()
print(end - start)
print(fails_dict)
```

```python
#Save the Data as a JSON file
api_df = []
with open('tweet_json.txt', 'r') as Json_file:
    for line in Json_file:
        tweet = (json.loads(line))
        tweet_id = tweet['id']
        retweet_count = tweet['retweet_count']
        favorite_count = tweet['favorite_count']
        create_date = tweet['created_at']
        api_df.append({'retweet_count' : retweet_count,
                  'favorite_count' : favorite_count,
                  'create_date' : create_date,
                  'tweet_id' : tweet_id})
```

```python
#load the data into a dataframe
df_twitterdata = pd.DataFrame(api_df, columns = ['tweet_id', 'retweet_count', 'favorite_count', 'create_date'])
```
